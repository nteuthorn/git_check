```
import pandas as pd
import seaborn as sns 
sns.set()
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```
houseprices = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
```

```
plt.figure(figsize = (11,5))
sns.distplot(houseprices['saleprice']);
```

```
plt.figure(figsize = (11,5))
sns.distplot(houseprices['lotarea']);
```

```
plt.figure(figsize = (11,5))
sns.boxplot(y="saleprice", x="overallcond", data=houseprices);
```

```
plt.figure(figsize = (11,5))
sns.boxplot(y="saleprice", x="overallcond", hue='centralair', data=houseprices);
```

```
plt.figure(figsize = (11,5))
sns.scatterplot(x="lotarea", y="saleprice", data=houseprices);
```
